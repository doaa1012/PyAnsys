```python
import numpy as np

Q = 1000  
freq = 1  
T_wave = 1 / freq  
omega = 2 * np.pi * freq  
t_end = 5 * T_wave  
room_temp = 25  
num_steps_per_wave = 100  
time_values = np.linspace(0, t_end, num_steps_per_wave , endpoint=False)  
cos_values = np.cos(omega * time_values)  
heat_flux = Q * (1 - cos_values)  
import matplotlib.pyplot as plt

plt.plot(time_values, heat_flux)
plt.xlabel('Time')
plt.ylabel('Heat Flux')
plt.title('Heat Flux vs. Time')
plt.grid(True)
plt.show()


```
